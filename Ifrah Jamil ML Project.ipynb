```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"C:\Users\LENOVO\Downloads\heart_disease.csv")
```

```python
#checking missing values
df.isnull().sum()
```

```python
#separating numerical and categorical columns 
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

print("Numerical columns:", num_cols)
print("Categorical columns:", cat_cols)
```

```python
#Handling Missing values 
#Mode imputation for categorical values
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

#Mean imputation for numerical values 
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

```

```python
#Verification of missing values 
df.isnull().sum()
```

```python
#separating target variable from x and storing it in y 
target = 'Heart Disease Status'
x = df.drop(target, axis=1)
y = df[target]

```

```python
#Feature selection
selected_features = [
    "Age",
    "Blood Pressure",
    "Cholesterol Level",
    "BMI",
    "Sleep Hours"
]
#all selected features are numerical 
x = df[selected_features]
y = df["Heart Disease Status"]
```

```python
#Target encoding 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)   # False=0, True=1
```

```python

from sklearn.model_selection import train_test_split
```

```python
#train-test split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y) 
#stratify=y keeps class balance 

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)

x_test_scaled = scaler.transform(x_test)
```

```python
#importing libraries for model training and evaluation
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle

```

```python
#Applying ML models 
#1. Logistic Regression 

lr = LogisticRegression(
    max_iter=1000,
    class_weight="balanced"
)

lr.fit(x_train_scaled, y_train)

y_pred_lr = lr.predict(x_test_scaled)

print(" Logistic Regression ")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("\nClassification Report:\n", classification_report(y_test, y_pred_lr))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))

```

```python
#2. Random Forest

rf = RandomForestClassifier(
    n_estimators=50,
    random_state=42,
    class_weight="balanced"
)

rf.fit(x_train_scaled, y_train)

y_pred_rf = rf.predict(x_test_scaled)

print("\nRandom Forest")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))

```

```python
print("\n===== Model Comparison =====")
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))

```

```python
print("Final model selected: Random Forest")
```
